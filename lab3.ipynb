```python
import matplotlib.pyplot as plt
import numpy as np

# Number of trials
# Probability of success
# Number of samples
data = np.random.binomial(20, 0.3, 1000)
plt.hist(data, bins=20, density=True, alpha=0.6, color='m')
plt.title("Binomial Distribution")
plt.xlabel("Number of Succeses")
plt.ylabel("Frequency")
plt.show





import matplotlib.pyplot as plt
import numpy as np

# Lower Boundry
# Upper Boundry
data = np.random.uniform(0, 10, 1000)
plt.hist(data, bins=20, density=True, alpha=0.6, color='c')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()






import matplotlib.pyplot as plt
import numpy as np

# Average value
data = np.random.exponential(scale=1.0, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()








import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

mean = np.mean(data)
std_dev = np.std(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, p, 'k', linewidth=2)

plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()








import matplotlib.pyplot as plt
import numpy as np

data = np.random.poisson (5, 1000)
plt.hist(data, bins=20, density=True, alpha=0.6, color='c')
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```
