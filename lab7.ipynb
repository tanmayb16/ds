```python
import numpy as np
import pandas as pd


df = pd.read_csv("C:/Users/borun/Desktop/diabetes.csv")



df.head(10)




df.columns




df





df.isnull().sum()






import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))
sns.histplot(df.Age)
plt.title('Age Distribution',size=18)
plt.xlabel('Age',size=15) #font size
plt.ylabel('Count',size=15)
plt.show()

sns.boxplot(df['Age'])
plt.title('Age Boxplot')
plt.show()







df['BMI'].hist(bins=20)
plt.title('BMI Distribution')
plt.show()









plt.figure(figsize = (10,6))
sns.heatmap(df.corr(),annot=True,square=True,
            cmap='RdBu',
            vmax=1,
            vmin=-1)
plt.title('Correlations Between Variables',size=18);
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()





df.shape[1]





sns.scatterplot(data=df, x='Glucose', y='BMI', hue='Outcome', palette='coolwarm', style='Outcome', s=100)
plt.title('Glucose vs BMI (Colored by Diabetes Outcome)')
plt.xlabel('Glucose Level')
plt.ylabel('BMI')
plt.legend(title='Diabetes Outcome')






sns.scatterplot(data = df)
plt.show()






sns.scatterplot(data = df, x='BMI', y='Age')
plt.show()






import itertools
for x, y in itertools.permutations(df.columns, 2):
    sns.scatterplot(data=df, x=x, y=y)
    plt.show()






```
